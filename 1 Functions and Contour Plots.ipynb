# Plotting univariate functions

Plotting of the function: $x^2-4x+4$

```python
# Define the function
def f(x):
    return x**2 - 4*x + 4

x = np.linspace(-6, 10, 400)

# Evaluate the function
y = f(x)

# Create the plot
plt.plot(x, y, label='$f(x) = x^2 - 4x + 7$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = x^2 - 4x + 7$')
plt.legend()


# Show the grid
#plt.grid(True)

# Show the plot
plt.show()
```

Plotting the function $\frac{1}{4} x^4 - \frac{1}{3} x^3 - x^2 +2$

```python
# Define the function
def f(x):
    return (x**4)/4 - (x**3)/3 - (x**2) + 2

x = np.linspace(-2, 3, 400)

# Evaluate the function
y = f(x)

# Create the plot
plt.plot(x, y, label='$f(x) = (1/4)x^4 - (1/3)x^3 - x^2 + 2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = (1/4)x^4 - (1/3)x^3 - x^2 + 2$')
plt.legend()


# Show the grid
#plt.grid(True)

# Show the plot
plt.show()
```

## Surface plots

$x^2+8y^2$

$x^2-y^2$

$\sqrt{x^2 + y^2}$

$e^{-x_1-x_2-1}+e^{-2x_1+2x_2-1}-e^{2x_1-x_2+1}$



```python
# Install plotly if not already installed
#!pip install plotly

# Import necessary libraries and configure Plotly for Colab
import plotly.graph_objs as go
import numpy as np
import plotly.io as pio
pio.renderers.default = 'colab'

# Define the function
def f(x, y):
    return np.sqrt(x**2 + y**2)

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute the corresponding z values
Z = f(X, Y)

# Create a surface plot
surface = go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')

# Define the layout
layout = go.Layout(
    title='Surface Plot of f(x, y)',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='f(x, y)'
    ),
    autosize=True
)

# Create the figure
fig = go.Figure(data=[surface], layout=layout)

# Plot the figure
fig.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the domain
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

# Nonconvex Function (Saddle Points): f(x, y) = x^2 - y^2
Z_nonconvex = X**2 - Y**2

# Convex Function (Global Minimum): f(x, y) = x^2 + y^2
Z_convex = np.sqrt(X**2 + Y**2)
# Function with Multiple Local Minima: f(x, y) = sin(x) * sin(y)
Z_multiple_minima = np.sin(X) * np.sin(Y)*Y*Y



# List of functions and their titles
functions = [(Z_nonconvex, 'Nonconvex Function: $f(x, y) = x^2 - y^2$'),
             (Z_convex, 'Convex Function: $f(x, y) = x^2 + y^2$'),
             (Z_multiple_minima, 'Function with Multiple Local Minima: $f(x, y) = \sin(x) \sin(y)$')]

for i, (Z, title) in enumerate(functions):
    fig = plt.figure(figsize=(12, 6))

    # Surface Plot with Contour Overlay
    ax1 = fig.add_subplot(121, projection='3d')
    surf1 = ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.8)
    ax1.contour3D(X, Y, Z, levels=10, cmap='coolwarm')
    ax1.set_title(title)
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_zlabel('f(x, y)')
    fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=5)

    # Contour Plot
    ax2 = fig.add_subplot(122)
    contour2 = ax2.contour(X, Y, Z, cmap='viridis', levels=10)
    ax2.set_title(f'Contour Plot of {title}')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    fig.colorbar(contour2, ax=ax2, shrink=0.5, aspect=5)

    plt.tight_layout()
    plt.show()
```

```python
# Install plotly if not already installed
#!pip install plotly

# Import necessary libraries and configure Plotly for Colab
import plotly.graph_objs as go
import numpy as np
import plotly.io as pio
pio.renderers.default = 'colab'

def f(x, y):
    return np.exp(x+3*y-0.1)+np.exp(x-3*y-0.1)+np.exp(-x-1)

# Create a grid of x and y values
x = np.linspace(-0.5, 0.5, 100)
y = np.linspace(-0.5, 0.5, 100)
X, Y = np.meshgrid(x, y)

# Compute the corresponding z values
Z = f(X, Y)

# Create a surface plot
surface = go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')

# Define the layout
layout = go.Layout(
    title='Surface Plot of f(x, y)',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='f(x, y)'
    ),
    autosize=True
)

# Create the figure
fig = go.Figure(data=[surface], layout=layout)

# Plot the figure
fig.show()

```
